```python
import pandas as pd

catspp_results = pd.read_csv('output/heuristics/catspp_results_20250626_021731.csv')
catspp_grasp_mejorado = pd.read_csv('output/heuristics/catspp_grasp_mejorado_results_20250626_021744.csv')
```

```python
import matplotlib.pyplot as plt

# Grfico de barras: comparacin de costos de grasp_mejorado vs grasp (catspp_results) vs bks
plt.figure(figsize=(14, 6))
plt.bar(catspp_grasp_mejorado['archivo'], catspp_grasp_mejorado['grasp_mejorado_costo'], label='GRASP Mejorado', alpha=0.7)
plt.bar(catspp_results['archivo'], catspp_results['grasp_costo'], label='GRASP', alpha=0.7)
plt.bar(catspp_grasp_mejorado['archivo'], catspp_grasp_mejorado['bks'], label='BKS', alpha=0.7)
plt.xticks(rotation=90)
plt.ylabel('Costo')
plt.title('Comparacin de Costo: GRASP Mejorado vs GRASP vs BKS')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 4))
plt.bar(catspp_results['archivo'], catspp_results['neighbour_costo'], label='Neighbour', alpha=0.7)
plt.xticks(rotation=90)
plt.ylabel('Costo')
plt.title('Costo Neighbour')
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 4))
plt.bar(catspp_results['archivo'], catspp_results['grasp_costo'], label='GRASP', alpha=0.7, color='orange')
plt.xticks(rotation=90)
plt.ylabel('Costo')
plt.title('Costo GRASP')
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 4))
plt.bar(catspp_results['archivo'], catspp_results['greedy_costo'], label='Greedy', alpha=0.7, color='green')
plt.xticks(rotation=90)
plt.ylabel('Costo')
plt.title('Costo Greedy')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Grfico de barras: comparacin de costos de grasp_mejorado vs bks
plt.figure(figsize=(12, 6))
plt.bar(catspp_grasp_mejorado['archivo'], catspp_grasp_mejorado['grasp_mejorado_costo'], label='GRASP Mejorado', alpha=0.7)
plt.bar(catspp_grasp_mejorado['archivo'], catspp_grasp_mejorado['bks'], label='BKS', alpha=0.7)
plt.xticks(rotation=90)
plt.ylabel('Costo')
plt.title('Comparacin de Costo: GRASP Mejorado vs BKS')
plt.legend()
plt.tight_layout()
plt.show()

# Grfico de barras: gap_percent de grasp_mejorado
plt.figure(figsize=(12, 6))
plt.bar(catspp_grasp_mejorado['archivo'], catspp_grasp_mejorado['gap_percent'])
plt.xticks(rotation=90)
plt.ylabel('Gap (%)')
plt.title('Gap Porcentual de GRASP Mejorado')
plt.tight_layout()
plt.show()
```
